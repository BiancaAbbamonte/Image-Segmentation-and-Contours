```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
imagem = cv2.imread("alfabeto.jpg")
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(image, interpolation="none")
plt.show()
image2 = image.copy()
#crop_img = img[y:y+h, x:x+w]
image2 = image[20:150,100:400]
plt.imshow(image2, interpolation="none", cmap="gray")
plt.show()
```

```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np

img_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)

image_lower_hsv = np.array([50, 60, 60])  
image_upper_hsv = np.array([80, 255, 255])


mask_hsv = cv2.inRange(img_hsv, image_lower_hsv, image_upper_hsv)

plt.subplot()
plt.imshow(img_rgb)
plt.show()
plt.subplot()
plt.imshow(mask_hsv, cmap="Greys_r", vmin=0, vmax=255)
plt.show()
```

```python
contornos, _ = cv2.findContours(mask_hsv, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# para desenhar o contorno primeiro faz uma copia da imagem 

mask_rgb = cv2.cvtColor(mask_hsv, cv2.COLOR_GRAY2RGB) 
contornos_img = mask_rgb.copy() # Cópia da máscara para ser desenhada "por cima"

cv2.drawContours(contornos_img, contornos, -1, [0, 255, 255], 3);

plt.figure(figsize=(8,6))
plt.imshow(contornos_img);

cnt = contornos[0]

M = cv2.moments(cnt)
print( M )



# Calculo das coordenadas do centro de massa

cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])

print("centro de massa na possição: ",cx, cy)

size = 3
color = (0, 80, 240)


cv2.line(contornos_img,(cx - size,cy),(cx + size,cy),color,2)
cv2.line(contornos_img,(cx,cy - size),(cx, cy + size),color,2)


font = cv2.FONT_HERSHEY_PLAIN
text = cy , cx
origem = (0,20)

cv2.putText(contornos_img, str(text), origem, font,1,(0, 80, 240),1,cv2.LINE_AA)

plt.imshow(contornos_img);
```
